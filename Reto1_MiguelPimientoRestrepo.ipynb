**RETO 1: Modelo Naive Bayes**

0. Se procede con la carga de las librerias de trabajo

```python
import numpy as np #Funciones numéricas transformación y analisis de datos
import pandas as pd #Conectar el sistema con las de datos
```

1. Se procede con la carga y organización de la base de datos

```python
nxl='/content/1. BD2_CarCrash_Int.xlsx' #Ruta de acceso BD
XDB=pd.read_excel(nxl,sheet_name=0) #DataFrame: Tabla de Datos

XDB = XDB.dropna() #Eliminando de plano los registros que poseen datos faltantes
XDB = XDB[XDB["County"] == "SAN FRANCISCO"] #Concentración en datos: condado de San Francisco
XDB["CrashType"] = pd.Categorical(XDB["CrashType"]).codes #Convertir a Valor Numérico

#Se cargan las variables de interes
XDB = XDB[['County','Weekday','ViolCat','ClearWeather','Month','Daylight','CrashType','Highway']]

XD = XDB[['Weekday','ViolCat','ClearWeather','Month','CrashType','Daylight']] #Aquí estan las variables de entrada
yd = XDB['Highway'] #Aquí esta la variable de referencia
```

2. Implementamos el Modelo

```python
from sklearn.naive_bayes import GaussianNB #Cargar GaussianNB

mnb=GaussianNB() #Aplicar Naive Bayes
mnb.fit(XD,yd) #Aplicar Naive Bayes teniendo en cuenta variables

#Se obtiene la info del modelo
ct=mnb.class_count_
print("Como están separados los datos:", ct)
print("El total de los datos es:", len(XD))
print("Los datos porcentualmente estan div:", ct/len(XD))
print(" ") #Espacio (para estética)

#Se obtiene la info de las variables (media y desviación est)
u=mnb.theta_
print("Las medias de las varibles:\n", u)
print(" ") #Espacio (para estética)
std=mnb.var_ #Esto es varianza
std=np.sqrt(std) #Esto es desviación estandar (sqrt es raiz)
print("La desviación estandar de las variables\n:", std)
print(" ") #Espacio (para estética)

#Se identifican los valores superiores e inferiores de los ingresos
print("EL limite superior de los valores:\n", u+std)
print(" ") #Espacio (para estética)
print("El limite inferior de los valores:\n", u-std)
print(" ") #Espacio (para estética)
```

**Matriz de Confusión**
3. Se procede con la evaluación (desempeño/exactitud del modelo)

```python
ydp=mnb.predict(XD) #Pronostico de los positivos y negativos

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(yd,ydp)
print(cm)
print(" ") #Espacio (para estética)

VN=cm[0,0];FP=cm[0,1];FN=cm[1,0];VP=cm[1,1]
Ex=(VN+VP)/(VN+FP+FN+VP)
print("La exactitud del modelo es:", Ex) #Exactitud/Desempeño General: (Verdadreo Negativo + Verdadero Positivo) / Total Datos | (+ usual)

TE=(FN+FP)/(VN+FP+FN+VP) #Tasa de Error
print("La tasa de error es:", TE)

Sen=(VP/(VP+FN)) #Sensibilidad: Comportamiento del modelo (1)
print("La sensibilidad es:", Sen)

Esp=(VN/(VN+FP)) #Especificidad: Comportamiento del modelo (0) | (+ usual)
print("La especificidad es:", Esp)

pre=(VP/(VP+FP)) #Precisión: Comportamiento general del modelo identificando los + (1) | (+ usual)
print("La presición es:", pre)

pren=(VN/(VN+FN)) #Precisión Negativa: Comportamiento general del modelo identificando los - (0)
print("La presición negativa es:", pren)
```
